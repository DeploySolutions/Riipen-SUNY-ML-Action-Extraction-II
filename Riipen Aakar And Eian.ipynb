```python
import pandas as pd
import re

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from rake_nltk import Rake
import PyPDF2
import textract
import os
from PyPDF2 import PdfFileReader

pdf_docs_path = os.path.join("PDF")
one_pdf_path = os.path.join(pdf_docs_path,"81363.pdf")


punctuations = ['(',')',';',':','[',']',',','.']
stop_words = stopwords.words('english')
keywords2=[]
phrases2=[]

with open(one_pdf_path, 'rb') as f:
    pdf = PdfFileReader(f)
    num_pages = pdf.numPages
    print("Number of pages in the document = ", num_pages)
    for i in range(num_pages):
        pageObj=pdf.getPage(i)
        print("Page Number = ", i)
        text = pageObj.extractText().split("  ")
        for j in range(len(text)):
            onePage=text[j].replace("\n","")
            print(onePage)
            
            r = Rake()
            r.extract_keywords_from_text(onePage)
            
            phrases = r.get_ranked_phrases_with_scores()
            phrases2.extend(phrases)
            
            lowcase = onePage.encode('ascii','ignore').lower() #Lowercasing each word
            tokens = word_tokenize(lowcase.decode("utf-8"))
            keywords = [word for word in tokens if not word in stop_words and not word in punctuations]
            keywords2.extend(keywords)

    f.close()
```
